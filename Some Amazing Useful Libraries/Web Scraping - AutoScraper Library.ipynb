```python
from autoscraper import AutoScraper
url = 'http://www.cs.cmu.edu/~mehrbod/RR/41955207.xml'
wanted_list = ["pauli68" , "07/28/2006" , "like eating in a funeral parlor!" , "2"]
scraper = AutoScraper()
result = scraper.build(url, wanted_list)
print(result)
```

```python
scraper.get_result_similar('http://www.cs.cmu.edu/~mehrbod/RR/41955207.xml' , grouped = True)
```

```python
scraper.set_rule_aliases({'rule_43o8' : 'name' , 'rule_cmvb' : 'date' , 'rule_d466' : 'date_duplicate' , 'rule_avyb' : 'comment' , 'rule_dx60' : 'rating'})
scraper.keep_rules(['rule_43o8' , 'rule_cmvb' , 'rule_d466' , 'rule_avyb' , 'rule_dx60'])
```

```python
df = scraper.get_result_similar('http://www.cs.cmu.edu/~mehrbod/RR/41955207.xml' , grouped = True , group_by_alias=True)
```

```python
import pandas as pd
df = pd.DataFrame(df)
```

```python
from autoscraper import AutoScraper
url = 'https://www.flipkart.com/search?q=iphone+11&sid=tyy%2C4io&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_1_5_na_na_ps&otracker1=AS_QueryStore_OrganicAutoSuggest_1_5_na_na_ps&as-pos=1&as-type=RECENT&suggestionId=iphone+11%7CMobiles&requestId=d346cd2f-6944-4f20-b10f-a1fb18ade701&as-searchtext=iphoe'
wanted_list = ["APPLE iPhone 11 (Black, 64 GB)" , "₹46,999"]
scraper = AutoScraper()
result1 = scraper.build(url, wanted_list)
print(result1)
```

```python
scraper.get_result_similar('https://www.flipkart.com/search?q=iphone+11&sid=tyy%2C4io&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_1_5_na_na_ps&otracker1=AS_QueryStore_OrganicAutoSuggest_1_5_na_na_ps&as-pos=1&as-type=RECENT&suggestionId=iphone+11%7CMobiles&requestId=d346cd2f-6944-4f20-b10f-a1fb18ade701&as-searchtext=iphoe' , grouped = True)
```
